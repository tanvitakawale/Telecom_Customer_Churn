```python
def get_age_group(Age):
    if Age <= 25:
        return 'Teenager'
    elif Age <= 45:
        return 'Young adult'
    elif Age <= 55:
        return 'middle-aged'
    else:
        return 'old'
```

```python
tele['Age_group']=tele['Age'].apply(get_age_group)
```

QUESTIONS

###Demographic Analysis

1)Find the average monthly charges made by each Age Group?

```python
avg_charges = tele.groupby('Age_group')['Monthly Charge'].mean()
print("Average Monthly Charges by",avg_charges)

plt.figure(figsize=(10,6))
plt.bar(avg_charges.index,avg_charges.values,color=['skyblue','pink','yellow','purple'],width=0.8,label=avg_charges.index)
plt.title('Average Customer Monthly Charges  by Age Group')
plt.xlabel('Age_group')
plt.ylabel('Monthly Charge')
plt.legend(loc='best')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

2)Find the average monthly GB download by each Age Group?

```python
average = tele.groupby('Age_group')[ 'Avg Monthly GB Download'].mean()
print("Average monthly GB Download by",average)

plt.figure(figsize=(10,6))
plt.bar(average.index,average.values,color=['green','red','yellow','brown'],width=0.5)
plt.title('Average monthly GB Download by Age Group')
plt.xlabel('Age_group')
plt.ylabel( 'Avg Monthly GB Download')
plt.xticks()
plt.show()
```

3)Find the average number of dependents made by each age_group?

```python
avg_dependents = tele.groupby('Age_group')['Number of Dependents'].count()
print("Average number of dependents by each ",avg_dependents)

plt.figure(figsize=(11,7))
plt.bar(avg_dependents.index,avg_dependents.values,color=['green','orange','pink','purple'],width=0.8)
plt.title('Average number of dependents by each age_group')
plt.xlabel('Age_group')
plt.ylabel('Number of Dependents')
plt.xticks()
plt.show()
```

4)Find the average age of customers who uses different payment method?

```python
total=tele.groupby("Payment Method")['Age'].mean()
print("Average age of customers who uses",total)


plt.figure(figsize=(11,7))
sns.boxplot(x='Payment Method', y='Age', data=tele)
plt.title('Average age of customer who uses payment method')
plt.xlabel('Payment Method',color='Red')
plt.ylabel('Age',color='Red')
plt.xticks()
plt.show()
```
